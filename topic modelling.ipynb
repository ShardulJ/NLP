```python
import nltk
nltk.download('stopwords')
```

```python
df = df.drop(['web-scraper-order','web-scraper-start-url','author','date','next','next-href'],1)
```

```python
df['rating'] = df['rating'].str.split()
```

```python
lest = []
for lst in df['rating']:
    rate = lst[0]
    lest.append(rate)
```

```python
df['rating'] = lest
```

```python
df['rating'] = df['rating'].astype(float)
```

```python
reviews = df['content']
```

```python
reviews = [re.sub(r'[^\w\s]','',str(item)) for item in reviews]
```

```python
stopwords = set(stopwords.words('english'))
```

```python
texts = [[word for word in documents.lower().split() if word not in stopwords]for documents in reviews]
```

```python
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
        
texts = [[token for token in text if frequency[token]>1]for text in texts]        
```

```python
dictionary = corpora.Dictionary(texts)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
NUM_OF_WORDS = 10
lda_model = LdaModel(corpus,num_topics = NUM_OF_WORDS,id2word = dictionary)
```

```python
topics = lda_model.show_topics()
```

```python
for topic in topics:
    print(topic)
```

```python
word_dict = {}
for i in range(NUM_OF_WORDS):
    words = lda_model.show_topic(i,topn=20)
    word_dict['topic#' + '{:02d}'.format(i+1)] = [word[0] for word in words]
pd.DataFrame(word_dict)
```
